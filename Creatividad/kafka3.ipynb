```python
# This module provides a portable way of using operating system dependent functionality.
import os

#The io module provides Python’s main facilities for dealing with various types of Input/Output
import io

#The glob module finds all the pathnames matching a specified pattern according to the rules used by the Unix shell
import glob
# La ruta donde están todos los archivos (este caso toma la ruta local)
work_dir = os.curdir

# Selecciona los archivos con el patrón indicado/aca habria q ver si quiero.
#Parsear.
for path in glob.glob(os.path.join(work_dir, "dramaturgia.txt")):
    with io.open(path, mode="r", encoding="latin-1") as fd:
        content = fd.read()
      
      
# Con esto podés chequear que estén todos       
# print(glob.glob("test*.txt"))

# NLTK is a leading platform for building Python programs to work with human language data.
from nltk import sent_tokenize

# separa el texto en oraciones
sentences = sent_tokenize(content)

# Para elegir un elemento al azar
import random

# import string
def RandomKafka(w):
    a = [ t for t in sentences if w in t]
    b = len(a)
    
    # Índice de la oración random seleccionada
    i = random.choice(range(len(a)))

    # Selecciona una oración random que contenga la palabra w
    # c = random.choice(a)
    
    # pasa todo a una única línea
    d = a[i].replace("\n"," ")
    
    # Quita todos los signos de puntuación    
    # e = d.translate(str.maketrans(" ", " ", string.punctuation)) 
    

    f = d.split(w)
    
    print(f[1])
    print()
    print("Resultado: " + str(i+1)+"/"+str(b))

    
    
w = input("Palabra buscada: ")
print()
RandomKafka(w)



    
```
